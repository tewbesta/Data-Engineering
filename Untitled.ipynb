```python
#d e
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = [2, 4, 6, 8, 10]
X=agg_df.iloc[:,1:]
print(X)
closest=[]
close={}
alldistances={}
for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(X)
    km=kmeanModel.fit(X)

    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / X.shape[0])
    inertias.append(kmeanModel.inertia_)
    d={}
    count=0



    mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / X.shape[0]
    mapping2[k] = kmeanModel.inertia_
    km = KMeans(n_clusters=k)
    alldistances[k] = km.fit_transform(X)
    for value in alldistances[k]:
            dist = np.linalg.norm(value[0] - value[1])
            d[count] = dist
            count+=1
    #print("the value of",d)
    sort_orders = sorted(d.items(), key=lambda x: x[1], reverse=True)
    temp=list(sort_orders)[0:3]
    print(temp)
    
    #e
    
    for i in range(3):
        print("the nearest neighbor artist for k=",str(k)+str(" ") +agg_df.index[temp[i][0]])
      #  print("the nearest neighbor genre for k=", str(k) + str(" ") + agg_df.co[temp[i][0]])

for key, val in mapping1.items():
    print("the centroids for each number of cluster is")
    print(f'{key} : {val}')
closest=[]

plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
#8 clusters are good because the elbow of the inertia curve flattens around that point and we want the minimum number of clusters with low inertia. 8 numbers of clusters provide that trade off well.
con.commit()
con.close()
```
